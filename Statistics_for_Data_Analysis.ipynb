### **Inferential Statistics**

Inferential statistics (also known as statistical inference) is a branch of statistics that deals with drawing conclusions or making predictions about a population based on a sample of data.

Descriptive statistics summarize and organize the data we already have (e.g., averages, graphs, or distributions), while inferential statistics allow us to go beyond the observed data to estimate, predict, or validate hypotheses about a larger group.

In this section, we will cover the fundamental principles of populations and sample data, and examine how distributions and their characteristics (e.g.median and standard deviation) can be used to draw inferential conclusions.



#### **Population and Samples**

Let's start by creating a simulation of a normal population data, and then, randomly choose a sample to compare their mean, median and mode. To do this, we will generate random data as follows:

```python
# Set seed for reproducibility
np.random.seed(1234)

# Create a normal population data (median = 100, standard deviation = 2, number of data points = 100.000)
population = np.random.normal(loc=100, scale=2, size=100000)

# Calculate mean, median and mode
mean_p = np.mean(population)
median_p = np.median(population)
mode_p = np.argmax(np.bincount(population.astype(int)))

# Density Curve
sns.kdeplot(population, fill=True, color='skyblue')
plt.title('Density Curve')
plt.xlabel('Value')
plt.ylabel('Density')


# mean, median and mode lines

plt.axvline(mean_p, color='red', linestyle='--', linewidth=1, label=f'Mean = {mean_p:.2f}')
plt.axvline(median_p, color='green', linestyle='-.', linewidth=1, label=f'Median = {median_p:.2f}')
plt.axvline(mode_p, color='purple', linestyle=':', linewidth=1, label=f'Mode = {mode_p:.2f}')

plt.legend()
plt.show()
```
